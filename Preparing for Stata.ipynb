```python
import pandas as pd
```

# Load data

```python
df = pd.read_parquet("df_cleaned_reg.parquet")
```

# Clean

```python
# drop unused column
df.drop(columns=['hammer_price_usd'], inplace=True)
```

```python
# drop empty rows
df = df.dropna(subset=[
    "auction_house_name",
    "auction_location",
    "medium_final",
    "artist_name",
])
```

```python
# consolidate auction houses
top50 = df["auction_house_name"].value_counts().nlargest(50).index

df.loc[:, "auction_house_name"] = df["auction_house_name"].where(
    df["auction_house_name"].isin(top50),
    "Other"
)
```

```python
# consolidate auction locations
top50 = df["auction_location"].value_counts().nlargest(50).index

df.loc[:, "auction_location"] = df["auction_location"].where(
    df["auction_location"].isin(top50),
    "Other"
)
```

```python
# consolidate currencies
top50 = df["auction_location"].value_counts().nlargest(50).index

df.loc[:, "auction_location"] = df["auction_location"].where(
    df["auction_location"].isin(top50),
    "Other"
)
```

```python
# consolidate artists
top50 = df["artist_name"].value_counts().nlargest(50).index

df.loc[:, "artist_name"] = df["artist_name"].where(
    df["artist_name"].isin(top50),
    "Other"
)
```

```python
# consolidate artist nationalities
top50 = df["artist_nationality"].value_counts().nlargest(50).index

df.loc[:, "artist_nationality"] = df["artist_nationality"].where(
    df["artist_nationality"].isin(top50),
    "Other"
)
```

# Export

```python
df.to_parquet("df_ols.parquet", index=False)
```
